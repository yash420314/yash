```python
##import libraries
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##reading the dataset
data = pd.read_csv('/Users/yashs/Downloads/Customer_Service_Requests_Analysis_Dataset/Customer_Service_Requests_Analysis_Dataset/311-service-requests-nyc/311_Service_Requests_from_2010_to_Present.csv',parse_dates = True, infer_datetime_format = True,low_memory = False )
df = pd.DataFrame(data)
```

```python
##Frequency of null values

plt.figure(figsize = (15,10))
null_count[null_count > 0].plot(kind = 'bar')
```

```python
df.drop([
"Intersection Street 1","Intersection Street 2","Landmark",
"School or Citywide Complaint","Vehicle Type","Taxi Company Borough","Taxi Pick Up Location",
"Bridge Highway Direction","Road Ramp","Garage Lot Name","Bridge Highway Segment","Ferry Terminal Name"],axis=1,inplace=True)
```

```python
null_count*100/len(df)
```

```python
## Closed Date column without null value
closed_date = df.dropna(subset=['Closed Date'])
print(closed_date)
```

```python
##Elapsed time
df['Closed Date'] = pd.to_datetime(df['Closed Date'])
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Elp_time'] = df['Closed Date'] - df['Created Date']
df['Elp_time']

## Elapsed time in sec
df['Elp_time'] = df['Elp_time'].apply(lambda x : x.seconds)
df['Elp_time']
```

```python
df.dtypes
```

```python
## Descriptive statistics of elapsed time

df['Elp_time'].describe()
```

```python
## Null values of city and complaint types 
City = pd.isnull(df['City']).sum()
print(City)
Complaint = pd.isnull(df['Complaint Type']).sum()
print(Complaint)
```

```python
## Impute the unknown city with value

df['City'].fillna(value = 'ajmer',inplace = True)
df['City'].head(40)
```

```python
##Complaints in each city

plt.figure(figsize=(15,10))
plt.title('Citywise complaints')
sns.countplot(y='City',data=df)
plt.show()
```

```python
Brooklyn = df[df['City'] == 'BROOKLYN']
Brooklyn[['Latitude','Longitude']].plot(kind = 'scatter',figsize = (12,10),x = 'Latitude',y = 'Longitude',title = 'Brooklyn')
```

```python
Brooklyn = df[df['City'] == 'BROOKLYN']
Brooklyn[['Latitude','Longitude']].plot(kind = 'hexbin',figsize = (12,10),x = 'Latitude',y = 'Longitude',title = 'Brooklyn',colormap = 'jet',gridsize= 35)
```

```python
##Top 10 complaints in New York City
df['Complaint Type'].value_counts().head(10)
```

```python
##Various types of complain in each city
Complaints = df.groupby(['City','Complaint Type'],as_index=False)['Unique Key'].count()
Complaints
```

```python
##df_new
df_new= pd.crosstab(df['Complaint Type'],df['City'])
df_new
```

```python
## Visualize type of complaints in each city
df.groupby(['City','Complaint Type'])['Unique Key'].size().unstack().plot(kind= 'bar',figsize=(15,10),
                                                               xlabel='City',ylabel='Number',
                                                               stacked=True,
                                                               title='Major complaint type in each city')
plt.show()
```

```python
grouped_data = df.groupby(['Complaint Type', 'Location'])['Elp_time'].mean().reset_index()
sorted_data = grouped_data.sort_values(by= 'Elp_time')
plt.figure(figsize=(15, 10))
plt.barh(sorted_data['Complaint Type'], sorted_data['Elp_time'])
plt.xlabel('Complaint Type')
plt.ylabel('Elp_time')
plt.title('Elapsed Time by Complaint Type by Location')
plt.show()
```
